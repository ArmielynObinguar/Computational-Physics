```python
#Python program of Lagrange's Interpolation
import math

# To represent a data point corresponding to x and y = f(x)
class Data:
 def __init__(self, x, y):
  self.x = x
  self.y = y

#Langrange interpolation  function to interpolate the given data points
def interpolate(f: list, xi: int, n: int) -> float:
 # Initialize result to 0
 result = 0.0
 for i in range(n):
  # Compute individual terms
  term = f[i].y
  for j in range(n):
   if j != i:
    term = term * (xi - f[j].x) / (f[i].x - f[j].x)
  # Add current term to result
  result += term
    #return the result
 return result

# Driver Code
if __name__ == "__main__":
    #5 known data points
    p=[Data(0.5,0.8415), Data(1.0,0.9093), Data(1.5,0.1411), Data(2,-0.7568), Data(2.5,-0.9589)]
    #find Lagrange interpolation at x=1.25
    res=interpolate(p,1.25,5)
    #compute f for sin2x, where x=1.25
    f=math.sin(2*1.25)
    #print the result
    print("Value of f(1.25) is :",res)
    print("Error at 1.25 is :",f-res)
```
